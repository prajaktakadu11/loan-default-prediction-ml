# Exploratory Data Analysis (EDA)

Check Target Distribution (Visual)

```python
sns.countplot(x='Default', data=df)
plt.title("Default Distribution")
plt.show()
```

Bivariate Analysis (Relation with Target)

```python
num_cols = ['Age','Income','LoanAmount','CreditScore','DTIRatio','InterestRate']

for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x='Default', y=col, data=df)
    plt.title(f"{col} vs Default")
    plt.show()
```

Distribution by Default (Advanced Insight)

```python
sns.kdeplot(data=df, x='Income', hue='Default', fill=True)
plt.title("Income Distribution by Default")
plt.show()
```
